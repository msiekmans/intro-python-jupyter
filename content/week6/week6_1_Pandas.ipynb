## 1. Wat is Pandas?

**Pandas** is de krachtigste bibliotheek in Python voor het werken met tabellen (**DataFrames**). Voor een geo-specialist is dit essentieel: waar een lijst slechts één dimensie heeft, kan Pandas duizenden rijen en kolommen aan informatie tegelijk verwerken.

### Waarom Pandas?
* **Snelheid:** Het verwerkt duizenden metingen in een fractie van een seconde.
* **Structuur:** Het houdt verbanden tussen data vast (bijv. een meting hoort bij een specifieke soort of locatie).
* **Voorbereiding:** Het is de directe opstap naar **GeoPandas**, de tool waarmee we later echte kaarten en ruimtelijke analyses gaan maken.

## 4. De Data Samenvatten (df.describe)

Nu de data is ingeladen, willen we niet elke rij apart gaan lezen. We willen een overzicht van de hele dataset. Hiervoor gebruiken we de functie `.describe()`.

### Wat doet .describe()?
Deze functie berekent automatisch de belangrijkste statistische waarden voor alle numerieke kolommen in je DataFrame:

* **count**: Het aantal ingevulde cellen (handig om te checken op missende data).
* **mean**: Het gemiddelde.
* **std**: De standaardafwijking (hoe ver liggen de metingen uit elkaar?).
* **min & max**: De kleinste en grootste waarden in de kolom.
* **25%, 50%, 75%**: De kwartielen (de mediaan en spreiding van de data).

### Waarom is dit nuttig voor Geo-Data?
Stel dat je een tabel hebt met de waterhoogtes van 100 meetpunten in Nederland. Met `.describe()` zie je in één seconde wat de gemiddelde waterstand is en of er uitschieters (extremen) tussen zitten die op een overstroming of defecte sensor kunnen duiden.



**Opdracht:** Voer de onderstaande cel uit en bekijk de tabel. Welk kenmerk heeft de grootste variatie (het grootste verschil tussen min en max)?

## 6. Correlatie: Verbanden ontdekken

Nu we kunnen filteren, willen we weten hoe de verschillende metingen met elkaar samenhangen. Dit noemen we **correlatie**.

### Wat is een Correlatiecoëfficiënt?
De waarde van een correlatie ligt altijd tussen de **-1** en **1**:
* **1**: Een perfect positief verband (als A groter wordt, wordt B ook groter).
* **0**: Totaal geen verband.
* **-1**: Een perfect negatief verband (als A groter wordt, wordt B juist kleiner).

### Waarom is dit belangrijk voor Geo-Data?
In de geo-wereld zoeken we vaak naar dit soort verbanden. Bijvoorbeeld: "Is er een correlatie tussen de nabijheid van een rivier en de prijs van een huis?" of "Hangt de bodemvochtigheid samen met de hoogte van het terrein?".

---

### Opdracht 6: De Correlatie Matrix

*Let op: Sinds kort moet je in Pandas aangeven dat je alleen de numerieke kolommen wilt gebruiken met `numeric_only=True`.*

```python
# Bereken de correlatie tussen alle metingen
correlatie_matrix = df.corr(numeric_only=True)

# Toon de matrix
correlatie_matrix
```

## 7. Visualisatie: Patronen Zichtbaar Maken

We hebben gezien dat `petal_length` en `petal_width` een extreem hoge correlatie hebben (bijna **0.96**). Maar wat betekent dat nou echt? De beste manier om dit te begrijpen is door een **Scatter Plot** te maken.

### Wat is een Scatter Plot?
Een scatter plot is een grafiek waarin we twee kenmerken op twee assen (x en y) tegen elkaar uitzetten. Elke stip in de grafiek is één bloem.

* **Clusteren:** Als we stippen dicht bij elkaar zien, noemen we dat een 'cluster'. Dit is een groep bloemen met vergelijkbare afmetingen.
* **Correlatie zien:** Omdat de correlatie hoog is, verwachten we een schuine lijn van stippen: als de lengte omhoog gaat, gaat de breedte ook omhoog.

### Waarom is dit belangrijk voor Geo-Data?
In de geo-wereld is een scatter plot vaak de eerste stap naar een kaart. De X en Y assen zijn dan de coördinaten (Longitude en Latitude). De kleur van de stip kan dan bijvoorbeeld de temperatuur of de bevolkingsdichtheid op die locatie weergeven. Hier gebruiken we de afmetingen en de kleur voor de soort.



---

### Opdracht 7: Maak de Scatter Plot
Gebruik de ingebouwde plot-functie van Pandas. We willen de kroonbladen vergelijken.

* **x:** `petal_length` (op de horizontale as)
* **y:** `petal_width` (op de verticale as)
* **c:** `species` (we gebruiken de soortnaam om de kleur van de stip te bepalen)

*Let op: In JupyterLite moet je soms een extra plot-bibliotheek importeren (`matplotlib.pyplot`) om de grafiek direct te tonen.*

```python
import matplotlib.pyplot as plt

# STAP 1: Zet de tekst-labels om naar getallen voor de kleurcode
color_labels = df['species'].astype('category').cat.codes

# STAP 2: Maak de scatter plot
scatter = df.plot(
    kind='scatter',
    x='petal_length',
    y='petal_width',
    c=color_labels,     # Gebruik de getallen voor de kleur
    colormap='viridis', # Een mooi kleurverloop
    colorbar=False, 
    figsize=(10, 6)
)

# STAP 3: Titels toevoegen voor de duidelijkheid
plt.title('Kroonblad Afmetingen per Soort (Iris Dataset)')
plt.xlabel('Kroonblad Lengte (cm)')
plt.ylabel('Kroonblad Breedte (cm)')

# Toon de grafiek
plt.show()
```

## 8. Wat vertelt deze Scatterplot ons nu?

De scatterplot is de "vertaling" van ruwe data naar visueel bewijs. In plaats van 150 rijen met getallen te scannen, zie je nu in één oogopslag de structuur van de natuur en de verschillen tussen de soorten.

### Duidelijke Groepsvorming (Clustering)
Je ziet dat de stippen niet willekeurig over de grafiek verspreid zijn. Ze vormen drie duidelijke groepen:
* **Iris Setosa (Paars):** Deze groep staat volledig apart linksonder. Dit vertelt ons dat de Setosa fysiek wezenlijk anders is dan de andere twee; er is geen overlap in de afmetingen van de kroonbladen.
* **Iris Versicolor & Virginica:** Deze groepen liggen dichter bij elkaar en overlappen een klein beetje. Dit vertelt ons dat een grote Versicolor soms precies dezelfde afmetingen kan hebben als een kleine Virginica.



### Bevestiging van de Correlatie
Omdat de stippen in een schuine lijn omhoog lopen (van linksonder naar rechtsboven), zie je de **positieve correlatie** die we eerder berekenden visueel terug:
* Het vertelt ons: "Hoe langer het kroonblad, hoe breder het kroonblad." 
* Er zijn vrijwel geen bloemen die heel lang zijn maar extreem smal, of heel breed maar extreem kort.

### Voorspellende Waarde
Dit is het belangrijkste doel van data-analyse. De plot vertelt ons dat we nu een **regel** kunnen opstellen: 
> *Stel: je vindt in het veld een Iris met een kroonblad van 1.5 cm lang. Zonder de bloem verder te bekijken, vertelt deze grafiek je dat het bijna 100% zeker een **Setosa** is.*

---

### De link naar Geo-Data
In jouw vakgebied vertelt een scatterplot (of een kaart met punten) vaak hetzelfde verhaal:
* **Clusters:** "Hier in deze wijk is de hittestress veel hoger dan in de rest van de stad."
* **Correlatie:** "Hoe meer verhard oppervlak (asfalt/beton) er in een buurt is, hoe hoger de temperatuur die we daar meten."



**De scatterplot maakt van 'losse metingen' dus een 'conclusie'.**

### Uitgebreidere uitwerking van bovenstaande voorbeeld
Hieronder nog een verder uitgewerkt voorbeeld van de scatterplot met legenda

```python
import matplotlib.pyplot as plt

# We maken een figuur aan
plt.figure(figsize=(10, 6))

# We lopen door de drie unieke soorten heen en plotten ze een voor een
for soort in df['species'].unique():
    subset = df[df['species'] == soort]
    plt.scatter(subset['petal_length'], subset['petal_width'], label=soort)

# Nu voegen we de extra informatie toe
plt.title('Afmetingen van Kroonbladen per Iris-soort')
plt.xlabel('Lengte Kroonblad (cm)')
plt.ylabel('Breedte Kroonblad (cm)')

# De belangrijkste stap: toon de legenda!
plt.legend(title='Bloemsoort')

# Raster toevoegen voor betere leesbaarheid
plt.grid(True, linestyle='--', alpha=0.3)

# Laat het resultaat zien
plt.show()
```
