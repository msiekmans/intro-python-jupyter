# Week1, Part 3: Working with Text (Strings)
In GIS and Data Science, we work extensively with names, categories, and labels. In Python, we call these **Strings**.

### Assignment 11: The Greeting
Create the variables `first_name` and `last_name`. Use an **f-string** to print: `"Welcome, [First Name] [Last Name]!"`

```python
first_name = "John"
last_name = "Doe"
print(f"Welcome, {first_name} {last_name}!")
```

### Assignment 12: Data Cleaning (Uppercase)
Create a variable `city = "utrecht"`. Use the `.upper()` method to print this as `UTRECHT`.

```python
city = "utrecht"
print(city.upper())
```

### Assignment 13: The Address Builder
Create `house_number = 10` (integer) and `street = "Dam"` (string). Use an f-string to combine them into a single variable named `full_address`.

```python
house_number = 10
street = "Dam"
full_address = f"{street} {house_number}"
print(full_address)
```

### Assignment 14: String Length
Use the `len()` function to count the number of characters in the string: `"Geographic Information Systems"`.

```python
text = "Geographic Information Systems"
length = len(text)
print(f"The text is {length} characters long.")
```

### Assignment 15: Searching in Text
Check if the word `"river"` occurs in the sentence: `"The Rhine is a major river in Europe"`. (The result should be True).

```python
sentence = "The Rhine is a major river in Europe"
check1 = "river" in sentence
check2 = "Netherlands" in sentence
print(check1)
print(check2)
```

## Interactive Programs
We use the `input()` function to allow the user to provide information to our program.

**Important:** `input()` always saves data as **Text (String)**. If you want to perform calculations with it, you must convert it using `int()` or `float()`.

### Assignment 16: The Welcome Bot
Ask the user for their name and print: `"Hello [name], welcome to the GIS lab!"`

```python
user_name = input("Please enter your name: ")
print(f"Hello {user_name}, welcome to the GIS lab!")
```

### Assignment 17: The Math Converter
Ask the user for a distance in **kilometers**. Convert this to a **float**, multiply it by 1000, and print the distance in **meters**.

```python
km_text = input("How many kilometers did you walk? ")
km_value = float(km_text)
meters = km_value * 1000
print(f"That is exactly {meters} meters.")
```

### Assignment 18: Professional Labels
Ask the user for a `city` and a `country`. Use an f-string to print them as an official map title: `"CITY, COUNTRY"` (e.g., "AMSTERDAM, NETHERLANDS").

*Hint: Use the .upper() method!*

```python
city = input("In which city are you now? ")
country = input("In what country is that? ")
title = f"{city.upper()}, {country.upper()}"
print(title)
```
