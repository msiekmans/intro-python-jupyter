## Assignment 3: Distance Approximation
**Goal:** Use the `*` operator.

**Question:** In geography, one degree of latitude is roughly **111 km**. Calculate how many kilometers are in **5.5 degrees** of latitude.

<details>

## Assignment 6: The PEMDAS Trap (Averages)
**Goal:** Understand why parentheses `()` are critical.

**Question:** Calculate the average of two latitude points: **52.1** and **52.9**. 

*Hint: You must add them first, THEN divide by 2. If you don't use parentheses, Python will divide the second number first!*

<details>
    <summary>Solution</summary>
    <code>(52.1 + 52.9) / 2</code>
</details>

## Assignment 10: Heron's Formula
**Goal:** Combining everything (Exponents, Parentheses, and Multiple Operations).

**Question:** Find the area of a triangle with sides **a=50**, **b=120**, and **c=130**.

**Step 1:** Calculate the semi-perimeter $s$:
$$s = (a + b + c) / 2$$

**Step 2:** Calculate the Area using Heron's Formula:
$$\text{Area} = \sqrt{s(s-a)(s-b)(s-c)}$$

*Hint: To do a square root in Python, raise the whole expression to the power of 0.5 using `** 0.5`.*

<details>
    <summary>Solution</summary>
    The calculation should result in: 3000 
</details>
