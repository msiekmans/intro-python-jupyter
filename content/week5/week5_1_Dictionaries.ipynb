# Week 5: Python Dictionaries voor Geo-data

## 1. Wat is een Dictionary?
Een **Dictionary** (letterlijk: woordenboek) is een verzameling waarin je gegevens opslaat in **sleutel-waarde paren** (*key-value pairs*). In de context van GeoMaps kun je dit zien als een digitale attributentabel waarbij elke rij een uniek label heeft.

* **De Sleutel (Key):** Het unieke label (bijv. de naam van een restaurant).
* **De Waarde (Value):** De informatie die bij die naam hoort (bijv. de beoordeling in sterren).

**Belangrijke regels:**
* Dictionaries staan tussen accolades `{ }`.
* Sleutels en waarden worden gescheiden door een dubbele punt `:`.
* Elk paar wordt gescheiden door een komma `,`.

---

## 2. Voorbeeld van een eenvoudige Dictionary
We beginnen met een basis-dictionary waarin we voor elk restaurant één gegeven (een getal) opslaan.


```python
# Eenvoudige Dictionary
# Sleutel: Naam | Waarde: Beoordeling
restaurant_scores = {
    "Sushi Sun": 4.8,
    "Burger Barn": 3.2,
    "Taco Town": 4.1,
    "Pasta Place": 3.9
}

# Gegevens opvragen via de sleutel
print(f"Beoordeling van Sushi Sun: {restaurant_scores['Sushi Sun']}")
```

## 3. Geneste Dictionaries

In de echte wereld hebben we vaak meerdere kenmerken nodig voor één locatie (zoals de keuken, prijs en sterren). Hiervoor gebruiken we een Geneste Dictionary (Nested Dictionary), waarbij de "waarde" zelf weer een dictionary is.

```python
# Geneste Dictionary
restaurants = {
    "Sushi Sun": {"stars": 4.8, "cuisine": "Japans", "price": "$$$"},
    "Burger Barn": {"stars": 3.2, "cuisine": "Amerikaans", "price": "$"},
    "Taco Town": {"stars": 4.1, "cuisine": "Mexicaans", "price": "$$"},
    "Pasta Place": {"stars": 3.9, "cuisine": "Italiaans", "price": "$$"}
}

# Specifieke gegevens opvragen (dieper graven)
print(f"Prijsklasse van Sushi Sun: {restaurants['Burger Barn']['price']}")
```

## 4. Loop en Print (Gegevens doorlopen)

Om elk restaurant in onze dictionary te "bezoeken", gebruiken we de methode `.items()`. Hiermee krijgen we tegelijkertijd toegang tot de **Naam** en de **Data-dictionary**.

```python
print("--- VOLLEDIGE RESTAURANTLIJST ---")

for naam, info in restaurants.items():
    sterren = info["stars"]
    keuken = info["cuisine"]
    print(f"Restaurant: {naam:12} | Keuken: {keuken:10} | Sterren: {sterren}")
```

## 5. Filteren naar een nieuwe Dictionary

We maken een gloednieuwe dictionary aan die alleen de restaurants bevat die aan een bepaalde voorwaarde voldoen (bijv. 4.0 sterren of hoger).

```python
# 1. Maak een lege dictionary voor onze selectie
top_selectie = {}

# 2. Doorloop de originele gegevens
for naam, info in restaurants.items():
    # 3. Controleer de voorwaarde
    if info["stars"] >= 4.0:
        # 4. Kopieer het hele record naar de nieuwe dictionary
        top_selectie[naam] = info

# 5. Bekijk het resultaat
print(f"Filter voltooid! \n{len(top_selectie)} top-locaties gevonden.")
print("-" * 45)
for naam, info in top_selectie.items():
    print(f"{naam:12} | Beoordeling: {info['stars']}")
```

# Waarom Dictionaries?

## Wat is GeoJSON?
**GeoJSON** is het meest gebruikte bestandsformaat voor webkaarten. Als je een kaart ziet op een website (zoals de kaart van Funda, Buienradar of een dashboard van de overheid), dan is de kans 99% dat de data als GeoJSON wordt verstuurd. Het is een tekst-gebaseerd formaat dat specifiek is ontworpen om geografische locaties (punten, lijnen, polygonen) én hun eigenschappen (naam, score, type) op te slaan.

## De Relatie met Dictionaries
In Python noemen we dit een **Dictionary**, maar in de wereld van data-uitwisseling noemen we dit **JSON** (*JavaScript Object Notation*). Ze zijn nagenoeg identiek.

Wanneer je een GeoJSON-bestand inlaadt in Python, verandert het automatisch in een dictionary. De hiërarchie van GeoJSON sluit naadloos aan op wat we zojuist hebben geleerd:

* **Dictionaries** bevatten de eigenschappen van een locatie (de zogenaamde "Properties").
* **Lists** (lijsten) bevatten vaak de coördinaten of een verzameling van verschillende locaties.

## Waarom is dit cruciaal voor GeoMaps & Data?
Zonder kennis van dictionaries kun je geen kaart data analyseren of visualiseren. Voor een data-student is dit om drie redenen essentieel:

1.  **Data-Extractie:** Je moet weten hoe je door de lagen van een dictionary navigeert om bijvoorbeeld alleen de "populatie" van een provincie op te halen uit een complex bestand.
2.  **Filtering:** Op een kaart wil je vaak niet alles tonen. Met dictionary-logica bepaal je welke data relevant is (bijv. "toon alleen de sensoren met een foutmelding").
3.  **Automatisering:** In plaats van handmatig punten te tekenen in GIS-software, gebruik je Python om duizenden datapunten in één keer te transformeren naar een interactieve kaart.
