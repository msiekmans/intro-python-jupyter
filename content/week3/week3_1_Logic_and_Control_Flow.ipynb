# Week 3: Comparison Operators & Control Flow

In this lab, we learn how to make Python "think." We use **Comparison Operators** to evaluate data and **If/Else statements** to perform different actions based on those results.

--- 

## 1. Logic & Comparison Operators
Comparison operators compare two values and return a **Boolean** (`True` or `False`). 

| Operator | Meaning | Example |
| :--- | :--- | :--- |
| `==` | Equal to | `city == "Amsterdam"` |
| `!=` | Not equal to | `status != "Error"` |
| `>` | Greater than | `elevation > 100` |
| `<` | Less than | `temp < 0` |
| `>=` | Greater or equal | `battery >= 20` |
| `<=` | Less or equal | `dist <= 500` |

--- 

## 2. If / Elif / Else (Control Flow)
This structure acts like a decision tree. 

**Crucial Rule:** Python uses **Indentation (4 spaces)** to know which code belongs inside the decision.

```python
dist_to_river = 450

if dist_to_river < 500:
    print("High Risk Zone")
elif dist_to_river <= 1000:
    print("Medium Risk Zone")
else:
    print("Safe Zone")
```
