### Het Verschil: `abs()` vs. `math.fabs()`

In Python zijn er twee manieren om de absolute waarde (de afstand tot nul) van een getal te berekenen. Hoewel ze hetzelfde doel hebben, verschillen ze in de manier waarop ze met data omgaan.

| Kenmerk            | `abs(x)`                          | `math.fabs(x)`                   |
| :----------------- | :-------------------------------- | :------------------------------- |
| **Bibliotheek**    | Ingebouwd (geen import nodig)     | Onderdeel van `math` module      |
| **Output Type**    | Behoudt type (int blijft int)     | Altijd een **float**             |
| **Complex getal**  | Ja (berekent de magnitude)        | Nee (geeft een `TypeError`)      |
| **Voorbeeld 1**    | `abs(-5)` &rarr; `5`              | `math.fabs(-5)` &rarr; `5.0`     |
| **Voorbeeld 2**    | `abs(-3.2)` &rarr; `3.2`          | `math.fabs(-3.2)` &rarr; `3.2`   |


#### Wanneer gebruik je welke?

1.  **`abs(x)`**: Dit is de standaardkeuze. Het is efficiënt en flexibel. Als je een heel getal invoert, krijg je een heel getal terug. Dit is meestal wat je wilt in Python.
2.  **`math.fabs(x)`**: De naam staat letterlijk voor "Float ABSolute". Gebruik deze alleen als je script strikt afhankelijk is van float-outputs voor verdere wiskundige berekeningen binnen de `math` bibliotheek.

> **Let op:** `math.abs()` bestaat niet in Python en zal een `AttributeError` veroorzaken. Gebruik altijd de ingebouwde `abs()` of `math.fabs()`.

## Het Afrondingsprobleem: Banker's Rounding

In Python (en de meeste moderne programmeertalen) werkt de standaard `round()` functie niet altijd zoals je op de basisschool hebt geleerd. Dit fenomeen heet **Banker's Rounding** (ook wel *Round Half to Even*).

### Wat is het probleem?
Bij het getal `.5` kijkt Python naar het **gehele getal vóór de punt**. 
* Is dat getal **even**? Dan rondt hij naar beneden af.
* Is dat getal **oneven**? Dan rondt hij naar boven af.

Kijk maar naar dit resultaat:

| Berekening | Resultaat | Waarom? |
| :--- | :--- | :--- |
| `round(2.5)` | **2** | 2 is even, dus hij blijft bij de 2. |
| `round(3.5)` | **4** | 3 is oneven, dus hij gaat naar de 4 (even). |

### Waarom doet Python dit?
Als je een lijst van 1.000.000 getallen hebt die allemaal op `.5` eindigen en je rondt ze allemaal naar boven af, dan wordt het totaalbedrag van je hele dataset veel hoger dan het in werkelijkheid is. Door de helft naar beneden en de helft naar boven af te ronden (naar het dichtstbijzijnde even getal), heffen de fouten elkaar op.

### School-rounding
Voor schoolwiskunde of financiële facturen willen we echter vaak de **"School-rounding"** regel: $\ge .5$ gaat altijd omhoog (verder van nul af).

| Waarde | `round()` (Banker) | `school_round()` (Verwacht) |
| :--- | :--- | :--- |
| `2.5` | 2 | **3** |
| `3.5` | 4 | **4** |
| `-2.5` | -2 | **-3** |
