## Functies en parameters

Informatie kan doorgegeven worden aan functies zodat hier een bewerking op uitgevoerd kan worden. We kunnen bijvoorbeeld een functie schrijven dat getallen verdubbelt.

```python
#voorbeeldprogramma: zelfgemaakte functie met parameters

def print_verdubbeling(x):
    y = 2*x
    print(x, "-->", y)

print_verdubbeling(12)
```

### return
Dit bovenstaande voorbeeld kunnen we mooier maken door de functie een waarde terug te laten sturen. Dit kan je doen met het woord **return**. We passen het bovenstaande voorbeeld een klein beetje aan:

```python
#voorbeeldprogramma: zelfgemaakte functie met return waarde

def bereken_verdubbeling(x):
    return 2*x

y = bereken_verdubbeling(12)
print(y)
```

Onderzoek voor jezelf wat er nu precies gebeurt met variabele **y** in de bovenstaande code.

We passen de code nog een beetje aan. Bestudeer het onderstaande programma. Wat is de uitkomst nu?

```python
#voorbeeldprogramma: zelfgemaakte functie met return waarde

def bereken_verdubbeling(x):
    return 2*x

y = bereken_verdubbeling(bereken_verdubbeling(12))
print(y)
```

## Libraries en functies
In een poging om Python code minder abstract te maken gaan we werken met wat visuele ondersteuning in de vorm van **ColabTurtlePlus** library.

Een **library** (bibliotheek) is een verzameling voor-geprogrammeerde code die je in je eigen project kunt laden, vergelijkbaar met een gereedschapskist vol kant-en-klaar gereedschap. Het dient om tijd te besparen en fouten te voorkomen; in plaats van dat je zelf complexe berekeningen of tekenfuncties vanaf nul moet schrijven, hergebruik je betrouwbare functies die door anderen zijn ontwikkeld en getest.

Om deze gereedschapskist te openen, gebruik je het commando `import`. Hiermee vertel je Python welke specifieke bibliotheek je wilt gaan gebruiken in je programma. Zodra een library is geïmporteerd, krijg je toegang tot alle functies die erin zitten, zodat je bijvoorbeeld met één simpel commando een schildpad kunt laten tekenen of geografische data kunt analyseren.

### ColabTurtlePlus library
Voer eerst de onderstaande **ColabTurtlePlus** code uit:

```python
#ColabTurtlePlus voorbeeld code
import piplite
await piplite.install('ColabTurtlePlus==2.0.1')

import warnings
warnings.filterwarnings("ignore", category=SyntaxWarning)

from ColabTurtlePlus.Turtle import *

clearscreen()
setup(300, 300)

t = Turtle()
t.pensize(3)
t.penup()
t.setposition(-50,0)
t.pendown()

# Draw something!
for i in range(4):
    t.forward(100)
    t.right(90);

t.color("red")

for i in range(3):
    t.forward(100)
    t.left(120)

#reset color
t.color("black")
t.hideturtle()
```

### Uitleg van de Turtle code
Deze Python code bereidt een digitaal tekenveld voor door de benodigde bibliotheek (**ColabTurtlePlus**) te installeren en storende systeemwaarschuwingen te negeren. De "schildpad" fungeert hier als een digitale plotter: via commando’s wordt de pen opgetild (`penup`), naar een specifieke startcoördinaat gestuurd (`setposition`) en weer op het papier gezet (`pendown`).  

De aansturing gebeurt door specifieke **functies** vanuit de bibliotheek aan te roepen, zoals `t.forward(100)` of `t.right(90)`. Hierbij fungeert `t` als het **object** (de schildpad) en vertelt de functienaam precies welke actie moet worden uitgevoerd. De waarde tussen de haakjes (het **argument**) bepaalt de exacte afstand of de hoek.  

Met behulp van **lussen** (loops) voert de schildpad deze functies herhaaldelijk uit: eerst ontstaat een zwart vierkant, waarna de pen naar rood wisselt voor een gelijkzijdige driehoek. Een belangrijk onderdeel van deze code is de functie `setup(300, 300)`, waarmee de afmetingen van het digitale tekenveld (het canvas) wordt bepaald. In dit geval creëert Python een venster van 300 bij 300 pixels, waarin men kan tekenen. Normaal begint de *Turtle* in het midden van het tekenveld te tekenen. Vandaar dat de `setposition` functie wordt aangeroepen om de tekenlocatie te wijzigen.

## Oplossing Opdracht: teken meerdere huizen via één functie

```python
def teken_huis(x, y, kleur):
    # 1. Ga naar de juiste startpositie (x, y) zonder te tekenen
    t.penup()
    t.goto(x, y)
    t.pendown()
    
    # 2. Teken de muren (vierkant)
    t.color(kleur)
    for i in range(4):
        t.forward(100)
        t.left(90)
        
    # 3. Teken het dak (driehoek)
    # Hint: De schildpad staat nu weer op (x, y). 
    # Hij moet eerst 100 omhoog om bij het dak te komen!
    t.penup()
    t.goto(x, y + 100) 
    t.pendown()
    
    t.color("red") # Dak is altijd rood
    for i in range(3):
        t.forward(100)
        t.left(120)

# --- TEST JE FUNCTIE HIERONDER ---
clearscreen()
setup(600, 300)
t = Turtle()
t.speed(6)
t.pensize(3)

# Teken nu drie verschillende huizen op verschillende plekken:
teken_huis(-150, -50, "blue")
teken_huis(0, -50, "orange")
teken_huis(150, -50, "green")

t.hideturtle()
```
