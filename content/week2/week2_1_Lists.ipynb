## 1. Lists (The Data Containers)
A **List** allows you to store multiple items in a single variable using square brackets `[]`. In Geomaps, we use these to store coordinates, city names, or elevation data.

### Key Concepts:
* **Creation:** `my_list = ["Item1", "Item2", "Item3"]`
* **Indexing:** Python starts counting at **0**.
    * `coords = [52.37, 4.89]` 
    * `coords[0]` is Latitude (52.37)
    * `coords[1]` is Longitude (4.89)
* **Adding Data:** Use `.append()` to add a new item to the end of a list.
    * `cities.append("Groningen")`
* **Counting:** Use `len(list_name)` to see how many items are in your dataset.

---

## 2. For Loops (The Automation Engine)
A **For Loop** repeats an action for every item in a list. Think of it as a **conveyor belt**: it picks up one item, performs a task, and automatically moves to the next until the list is empty.

### The Basic Syntax:
```python
cities = ["Amsterdam", "Rotterdam", "Utrecht"]

for city in cities:
    # Everything indented here runs for each city
    print(f"Mapping the city of: {city}")
```

---

## 3. The range() Function

The **range()** function generates a sequence of numbers without needing a pre-made list.  
```python
range(1, 6) 
```
This generates: 1, 2, 3, 4, 5 (it stops before the final number).  
**Usage:** Great for creating unique ID numbers for map markers or repeating a calculation a set number of times.

---

## 4. List of Lists
we can analyze a "List of Lists" (a route):


### The Syntax:
```python
# A list of [Lat, Lon, Alt]
route = [
    [52.1, 4.5, 10], 
    [52.2, 4.6, 130], 
    [52.3, 4.7, 20]
]

for point in route:
    altitude = point[2]  # Access the 3rd item (index 2)
    if altitude > 120:
        print(f"Alert: Altitude {altitude}m exceeds legal limit!")

```
